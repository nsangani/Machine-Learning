```python
# Libraries

import pandas as pd
import sklearn.linear_model as lm
import numpy as np 
```

```python
# Import data 
df = pd.read_csv("patient1.csv") # status and age 
# print(df.describe) # 167 rows
df1 = df.to_numpy()
#df1
```

```python
# create logistic regression model
x = df1[:,1:2] # age
y = df1[:,0] #status 

lr = lm.LogisticRegression()
lr.fit(x,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
# plot the logistic function
import matplotlib.pyplot as plt

plt.scatter(df1[:,1],df1[:,0]) 
test = np.arange(100).reshape((100,1))
res = lr.predict_proba(test)
plt.scatter(test[:,0],res[:,1], c = 'r')

plt.savefig('lr_function.png')
```

```python
# Question #4

import pandas as pd
import numpy as np
import keras
import tensorflow as tf
import keras.models as km
import keras.layers as kl
from tensorflow.keras.utils import plot_model

train_data = pd.read_csv("Immunotherapy_train.csv").to_numpy()
test_data = pd.read_csv("Immunotherapy_test.csv").to_numpy()
```

```python
# Build the model and save the model plot

model = km.Sequential()

# three dense layer
model.add(kl.Dense(32, input_dim = 7, use_bias = True, bias_initializer = "ones",
activation = "tanh"))
model.add(kl.Dense(4, use_bias = True, bias_initializer = "ones", activation =
"tanh"))
model.add(kl.Dense(1, use_bias = True, bias_initializer = "ones", activation =
"tanh"))

model.compile(loss="binary_crossentropy", optimizer = "rmsprop",
metrics=["accuracy"])

# tf.keras.utils.plot_model(model, to_file='model_Q4.png', show_shapes=False, show_dtype=False,show_layer_names=True, rankdir='TB', expand_nested=False, dpi=96)
```

```python
# Test and predict

train_x = train_data[:,0:7]
train_y = train_data[:,7]

model.fit(train_x, train_y, epochs=1000)
model.evaluate(train_x, train_y)

test_x = test_data[:,0:7]
test_y = test_data[:,7]

pred = model.predict(test_x)
print(pred)
```
